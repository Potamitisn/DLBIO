```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data loading (replace with your own dataframes)
df1 = pd.read_csv('maml_lr_finetuning.csv',index_col=0)
df2 = pd.read_csv('maml_dim_finetuning.csv', index_col=0)

# Create a figure with 2 subplots
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# First plot on the first subplot
sns.lineplot(data=df1.drop(columns=['loss']),  markers=True, ax=axs[0], dashes=False)  
axs[0].set_xlabel("Learning Rate")
# Second plot on the second subplot
sns.lineplot(data=df2.drop(columns=['loss']),  markers=True, ax=axs[1], dashes=False)  
axs[1].set_xlabel("Dim")
# Adjust layout
plt.tight_layout()
# Save it 
plt.savefig("maml_finetuning.png", bbox_inches='tight' )
# Show the plot
plt.show()
```
